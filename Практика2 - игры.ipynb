```python

#1 - Игра "Быки и коровы"
#2 - Игра "Wordle/5 букв"
#3 - Игра "Камень, ножницы, бумага, ящерица, Спок"
#0 - Выход из программы

from random import sample, randint
import re

def bulls_cows():
    digits = list('0123456789')
    first = sample(digits[1:], 1)
    digits.remove(first[0])
    rand_num = ''.join(first + sample(digits, 3)) 

    cows = bulls = count1 = 0
    while (cows != 4) and (count1 < 6):
        cows = bulls = 0
        try:
            your_num = input("Введите 4-значное число (Цифры в нем не должны повторяться!): ")
            if (len(your_num)!=4)or(len(set(your_num)) != len(your_num)):
                raise ValueError()
        except ValueError:
            print("Это число не подходит. Попробуйте снова...")
            print('---\n')
        else: 
            count1+=1
            print('///', rand_num)
            print (count1, '. Ваше число: ', your_num, sep='')
            for i in range(4):
                if rand_num[i] == your_num[i]:
                    cows += 1
                elif your_num[i] in rand_num:
                    bulls += 1
            print('Оно содержит', cows, 'коров(ы) и', bulls, 'бык(ов/а).')
    if cows == 4: print('Вы выиграли! Количество попыток -', count1)
    elif count1 > 5: print('Превышено количество попыток - 6. Вы проиграли(')
    
def is_russian(s: str) -> bool:
    return bool(re.search('[а-яА-Я]', s))

def wordle():
    words = ['озеро', 'балет', 'капри', 'микро', 'добро', 'укроп', 'цапля', 'ясень', 'ябеда', 'жизнь', 'кровь', 'загар', 'дрова', 
               'бровь', 'тапки', 'тапок', 'рулет', 'омлет', 'атлет', 'Ивлев', 'зверь', 'птица', 'выдра', 'лебедь', 'экран', 'напев', 
               'шерсть', 'шорох', 'порох', 'ересь', 'чуйка', 'чайка', 'сокол', 'покой', 'кусок', 'леска', 'ласка', 'лепет', 'горох', 
               'норов', 'боров', 'билет', 'лейка', 'рейка', 'акула', 'оброк', 'масть', 'пасть', 'пират', 'валет', 'актер', 'короб', 
               'жираф', 'франк', 'жрица', 'жердь', 'желчь', 'полет', 'маска', 'паста', 'рукав', 'опека', 'игрок', 'турок', 'хурма', 
               'выпрь', 'амеба', 'афера', 'белье', 'башня', 'взлет', 'заряд', 'индюк', 'петух', 'капля', 'карта', 'ладья', 'копье']
    rand_word = sample(words, 1)
    y = count2 = 0
    while (y != 5) and (count2 < 10):
        try:
            your_word = input("Введите слово из 5 букв (русского алфавита): ")
            if not is_russian(your_word):
                raise ValueError()
        except ValueError:
                print("Такого слова нет. Попробуйте снова...")
                print('---\n')
        else:
            count2+=1
            rand_word, your_word = ''.join(rand_word), ''.join(your_word)
            new_word = list(your_word)
            print('///', rand_word)
            print (count2, '. Ваше слово: ', your_word, sep='', end = '->')
    
            for i in range(len(your_word)):
                if your_word[i] == rand_word[i]:
                    new_word[i] = '[' + your_word[i] + ']'
                    y += 1  
                elif your_word[i] in rand_word:
                    new_word[i] = '(' + your_word[i] + ')'  
            print(new_word) 
    if y == 5: print('Вы выиграли! Количество попыток -', count2)
    elif count2 > 9: print('Превышено количество попыток - 10. Вы проиграли(')

def Spock():
    list_choice = ['камень', 'ящерица', 'Спок', 'ножницы', 'бумага']
    player1 = randint(0, 4)
    player2 = int(input("Ваш ход (введите номер): 0 - камень, 1 - ящерица, 2 - Спок, 3 - ножницы, 4 - бумага"))
    print('Ход противника:', list_choice[player1])
    print('Ваш ход:', list_choice[player2])
    if (player1-player2) == 0:
        print('Ничья!')
    elif (player1-player2) % 2 == 0:
        if (player1-player2) > 0:
            print('Вы проиграли!')
        else:
            print('Победа за Вами!')
    else:
        if (player1-player2) > 0:
            print('Победа за Вами!')
        else:
            print('Вы проиграли!') 

def default():
    q = input('Сыграем еще? д/н' )
    #if q == "н": quit() 

switch = {
    1: bulls_cows,
    2: wordle,
    3: Spock
}
choice = int(input())
switch.get(choice, default)()
```
